# Importing

```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(precision=8)
```

# Sigmoid Function

```python
def sigmoid(z):
    g = 1/(1 + np.exp(-z))
    return g
```

# Regularized Linear Cost Function

```python
def regularized_linear_cost(X,y,w,b,lambda_= 1):

    m,n = X.shape
    cost = 0
    for i in range(m):
        f_wbi = w @ X[i] + b
        cost += (f_wbi - y[i])**2
    cost = cost/(2*m)
    
    reg_cost = 0
    for j in range(n):
        reg_cost += w[j]**2
    reg_cost = (lambda_*reg_cost)/(2*m)

    j_wb = cost + reg_cost
    return j_wb
```

```python
np.random.seed(1)
X_tmp = np.random.rand(5,6)
y_tmp = np.array([0,1,0,1,0])
w_tmp = np.random.rand(X_tmp.shape[1]).reshape(-1,)-0.5
b_tmp = 0.5
lambda_tmp = 0.7
cost_tmp = regularized_linear_cost(X_tmp, y_tmp, w_tmp, b_tmp, lambda_tmp)

print("Regularized cost:", cost_tmp)
```

# Regularized Logistic Cost Function

```python
def reg_logistic_cost(X,y,w,b,lambda_):
    m,n = X.shape

    cost = 0
    for i in range(m):
        z = w @ X[i] + b
        f_wbi = sigmoid(z)
        cost1 = -y[i] * np.log(f_wbi)
        cost2 = -(1-y[i]) * np.log(1-f_wbi)
        cost += cost1 + cost2
    cost = cost/m

    reg_cost = 0
    for j in range(n):
        reg_cost = w[j]**2
    reg_cost = reg_cost*lambda_/(2*m)

    return cost + reg_cost
```

```python
np.random.seed(1)
X_tmp = np.random.rand(5,6)
y_tmp = np.array([0,1,0,1,0])
w_tmp = np.random.rand(X_tmp.shape[1]).reshape(-1,)-0.5
b_tmp = 0.5
lambda_tmp = 0.7
cost_tmp = reg_logistic_cost(X_tmp, y_tmp, w_tmp, b_tmp, lambda_tmp)

print("Regularized cost:", cost_tmp)
```
